```python
import cv2
from egbis.utils.graph import graph_partition_from_image, ndarray_partition_from_image
from egbis.segment import segment
import matplotlib.pyplot as plt
from egbis.visualize import draw_segmentation

big_image = cv2.imread("example/example_2.jpg")
small_image = cv2.resize(big_image, (256, 256))

ndarray_partition_from_image(small_image)
```

```python
result = graph_partition_from_image(small_image, directions=[(0, 1), (1, 0), (1, 1)])

result = segment(
    small_image,
    iterations=10,
    k=1000
)
```

```python
axes = plt.subplot()
draw_segmentation(axes=axes, segmentation=result, image=small_image / 255, opacity=0.9)
```

```python
axes = plt.subplot()
draw_segmentation(axes=axes, segmentation=result, image=small_image / 255, opacity=0.5)
```

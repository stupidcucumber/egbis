```python
import cv2
import matplotlib.pyplot as plt
from egbis.visualize import draw_components_partition
from egbis.segment import segment

image = cv2.imread("example/images/example_2.jpg")
image = cv2.resize(image, (256, 256))
blurred_image = cv2.GaussianBlur(image, (3, 3), 0.8)

result = segment(image, directions=[(1, 0), (0, 1)], k=20)
fig, axes = plt.subplots(ncols=2)

draw_components_partition(image, result, axes[0], 1.0)
draw_components_partition(image, result, axes[1], 0.0)
```
